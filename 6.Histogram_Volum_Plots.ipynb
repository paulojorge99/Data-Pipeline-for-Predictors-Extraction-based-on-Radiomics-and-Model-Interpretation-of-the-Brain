# PLOT VOLUMES

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
file1 = "0-segments_final.csv"
csv1 = pd.read_csv(file1)
print(csv1.shape)

#normalize the data

columns=["Paciente"]

def normalize_radiomic(dff):
    
    data = dff.drop(df.columns[0], axis=1)
    data = dff.drop(columns=columns)
        
    data_normalized = data.apply(lambda x:( (x - x.min()) / (x.max()-x.min())))
    
    return data_normalized
   
df=normalize_radiomic(csv1)

listaID=csv1["Paciente"]

df.insert(0, 'Paciente', listaID)

nan_value = float("NaN")
df.replace("", nan_value, inplace=True)
  
df.dropna(how='all', axis=1, inplace=True)

print(df.shape)
#df.to_csv("01-segments_final_normalized.csv", index = False)
```

```python
df = pd.read_csv("01-segments_final_normalized.csv")

lista=[]

for col in df.columns:
    lista.append(col)
lista.pop(0)    

for coluna in lista:
    #volume = df[coluna]
    if coluna == "Left-Hippocampus_volume" or coluna == "Right-Hippocampus_volume" :
        
        volume = df[coluna]
        
        plt.figure(figsize=(12, 6))
        plt.hist(volume)
        plt.xlabel(coluna)
        plt.ylabel('Pacientes')
        plt.show()
        plt.close()
```
